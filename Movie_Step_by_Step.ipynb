```python
import pandas as pd
```

```python
movies = pd.read_csv("./ml-25m/movies.csv")
movies
```

```python
ratings = pd.read_csv("./ml-25m/ratings.csv")
ratings
```

```python
movies_ratings = pd.merge(ratings, movies[['movieId', 'title']], on='movieId')
movies_ratings
```

```python
popularity = movies_ratings.groupby('title').size().reset_index(name='Most Popular')
popularity
```

```python
popularity = popularity.sort_values(by=['Most Popular'])
popularity
```

```python
ratings_mean = pd.DataFrame(movies_ratings.groupby('title')['rating'].mean())
ratings_mean
```

```python
ratings_mean = ratings_mean.sort_values('rating')
ratings_mean
```

```python
mean_popular = pd.merge(ratings_mean, popularity[['title', 'Most Popular']], on='title')
mean_popular
```

```python
mean_popular = mean_popular[mean_popular['Most Popular']>30000]
mean_popular
```

```python
ratings_mean = ratings_mean.reset_index()
userid_counts = pd.merge(movies_ratings, ratings_mean[['title','rating_counts']], on='title')
userid_counts
```

```python
user_ratings = userid_counts.drop(userid_counts[userid_counts.	rating_counts < 10000].index)
user_ratings
```

```python
user_movie_rating = user_ratings.pivot_table(index='userId', columns=['title'], values='rating')
user_movie_rating.head()
```

```python
toy_story_ratings = user_movie_rating['Toy Story (1995)']
movies_like_toy_story = user_movie_rating.corrwith(toy_story_ratings)
movies_like_toy_story
```

```python
corr_heading = pd.DataFrame(movies_like_toy_story, columns=['Correlation'])
recommended = corr_heading.sort_values(['Correlation'], ascending=False)
recommended
```
